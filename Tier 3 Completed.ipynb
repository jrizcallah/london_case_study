```python
# Let's check out some histograms of the boroughs of interest and see what we can learn from them.
plt.clf()
for x in ['City of London', 'Hackney', 'Bexley'] :
    temp = bor_pct_change[x]
    temp.hist(range=[-0.1,0.1])
    plt.axvline(temp.mean(), color='r', linestyle='dashed', linewidth=1)
    plt.axvline(temp.median(), color='k', linestyle='dashed', linewidth=1)
    plt.title(x)
    plt.savefig(temp.name+'.png')
    plt.show()
```

Bexley and City of London are symmetrically distributed, but Hackney's percentage changes are skewed to the left, meaning there is a tail of low values dragging down the mean.

```python
# Now we can explore the yearly percent changes.
bor_yr_change = boroughs.resample('y', label='left', loffset='1D').first().pct_change()
bor_yr_change.head()
```

```python
temp = 1 + bor_pct_change
total_pct_change = np.cumprod(temp).iloc[-1,:].sort_values(ascending=False)
print('The cumulative percent change in housing price in each borough is: \n')
print(total_pct_change)
```

```python
temp = 1 + bor_yr_change
total_yr_change = np.cumprod(temp).iloc[-1,:].sort_values(ascending=False)
print('The cumulative yearly percent change in housing price in each borough is: \n')
print(total_yr_change)
```

```python
numdays = prop_data.index[-1] - prop_data.index[0] 
numdays = numdays.total_seconds()/(60*60*24)
numyears = numdays/365
nummonths = numyears*12
numyears = numyears // 1
```

```python
avg_pct_change = pd.DataFrame({'Yearly':total_yr_change, 'Monthly':total_pct_change})
avg_pct_change['Yearly'] = ((avg_pct_change['Yearly']**(1/numyears) - 1) * 100).round(2)
avg_pct_change['Monthly'] = ((avg_pct_change['Monthly']**(1/nummonths) - 1) * 100).round(2)
avg_pct_change = avg_pct_change.sort_values('Yearly', ascending=False)
print('The Average Percent Change in each borough from January 1995 to 2020 is: \n')
print(avg_pct_change)
```

```python
plt.clf()
plt.title('Average Yearly Percent Change')
plt.xticks(rotation=90)
plt.vlines(x=avg_pct_change.index, ymin=0, ymax=avg_pct_change['Yearly'], color='#007acc', alpha=0.2, linewidth=5)
plt.plot(avg_pct_change.index, avg_pct_change['Yearly'], "o", markersize=5, color='#007acc', alpha=0.6)
plt.ylabel('Percentage', fontsize=15, fontweight='black', color = '#333F4B')
plt.xlabel('')
plt.savefig('yearly_change.png')
plt.show()
```

```python
plt.clf()
avg_pct_monthly = avg_pct_change.sort_values('Monthly', ascending=False)
plt.title('Average Monthly Percent Change')
plt.xticks(rotation=90)
plt.vlines(x=avg_pct_monthly.index, ymin=0, ymax=avg_pct_monthly['Monthly'], color='#007acc', alpha=0.2, linewidth=5)
plt.plot(avg_pct_monthly.index, avg_pct_monthly['Monthly'], "o", markersize=5, color='#007acc', alpha=0.6)
plt.ylabel('Percentage', fontsize=15, fontweight='black', color = '#333F4B')
plt.xlabel('')
plt.savefig('monthly_change.png')
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
# This function takes in the name of a borough, then calculates the ratio of the housing price in 2020 to the housing
# price in 2000 (twenty years ago)
def house_ratio(Area):
    temp = pd.DataFrame(prop_data[Area], prop_data.index)
    temp['year'] = temp.index.year
    temp[Area] = temp[Area].astype(float)
    temp2 = temp.groupby('year').mean()
    out = temp2.loc[2020] / temp2.loc[2000]
    print('An average house in ' + Area + ' is worth ' + str(round(out[0], 2)) + ' times as much in 2020 as it was in 2000')
    
```
